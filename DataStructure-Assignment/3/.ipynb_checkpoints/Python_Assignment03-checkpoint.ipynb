# Question 1 Algorithm
Step 1: Please enter the file name with words and meanings 

Step 2: The program picks a random word from the file and scrabmles the word.

# Note: Random module is used to pick the random word from the file and shuffle is used to scramble the word 

A Dictonary is used to keep the word and the meaning in the form of Key- Value pair, using which the program picks the random key and scrambles the word and when the user asks for the hint , the program picks the correspoinding value and prints it inorder to help the user guess the right word. 

Step 3: The user takes the Guess. If the Entered word matches with the key, then the program asks the user if he/she wants to continue the game or exit the game. 

Step 4: If thr user decides to continue the game and enters the "yes", then the program repeats the process of picking a random word from the file and and scrambling it and asking the user to guess the word. 

Step 5: If the user is unable to guess the word and he requires a Hint, then he enter "?" asking the program to give a Hint. During which the program picks the corresponding value and shows the user the meaning of the word. If it matches then the process continues.

Step 6: If the word does not match then the user is provided with n number of chances until he guesses the right answer. 

Step 7: If the user guesses the right answer as he is asked if he wants to continue or not, if he decides not to continue, he types "no" and the program quits with a "GoodBye!" Message. 

Step 8: If he types anything other than "yes" or "no" then the program Quits saying that the user has not given a required input. 

# Please note the program gives all the meanigs of the words when asked for a hint.

```python
#Question 1
from random import choice
from random import shuffle
newDict = {}
val = 'yes'
try:
    filename = input("Give the name of the “words and their meanings” file: ")
    fhandle = open(filename,'r')
    for line in fhandle:
        words = line.split(',')
        word1 = ''.join(words[1:])
        newDict[words[0]] = word1.strip()
    
    while(val == 'yes'):
        myword = choice(list(newDict.keys()))
        my_string = list(myword)
        shuffle(my_string)
        scrambled = ''.join(my_string)
        print("Unscramble the following letters to form a word. Type “?” for the meaning of the unscrambled word: ",scrambled)
        answer = input("Enter the answer [or ? for the meaning]: ")
        while(answer != myword):
            if answer == '?':
                print('The word means: ',newDict[myword])
            else:
                print("Wrong, try again")
            answer = input("Enter the answer [or ? for the meaning]: ")
        prompt = str(input("You got it! Do you want to continue [yes or no]: "))
        val = prompt
    if(val == 'no'):
        print('Goodbye!')  
    else:
        print("Please provide only yes or no as inputs")
except:
    print('Please provide a valid filename')
```

# Question 2 Algorithm
Step 1: Retrieve file name as input
Step 2. Open and iterate through each line of the file
Step 3. In each iteration , splits words in each line and store it in variable content
Step 4. store first word of content in variable name which will have all the names of reviewers
Step 5. create an empty dictionary named newDict
Step 6. Iterate for i in range(1,len(content),2)
Step 7. With each iteration, store content[i] in variable read_name
Step 8. Store content[i+1] in variable read_review
Step 9. Store read_review as value to newDict[read_name]
Step 10. Create another empty Dictionary largeDict
Step 11. Store newDict to largeDict[name]
Step 12. Prompt user to enter input 1, 2 or 3 for associated operations
Step 13. If the user enters 1, retrieve names of 2 reviewers as input
Step 14. call the similarityScore(reviewer1,reviewer2) function.
Step 15. This function definition will take reviewer1 and reviewer2 as parameters, perform Euclidean distance by the formula
Step 16. Perform square root of the result using math.sqrt(sum)
Step 17. If the user enters 2, retrieve 1 reviewer name as input as store it in review1
Step 18. Iterate using for review2 in largeDict.keys() and check if review1 == review2
Step 19. If equal, then continue to skip calculation for same values
Step 20. If not equal, call the similarityScore(review1,review2) function
Step 21. If user enters 3, print statement 'Goodbye!' and break out of the loop
Step 22. Exit the program

```python
#Question 2
import math

largeDict = {}

def similarityScore(reviewer1,reviewer2):
    if reviewer1 in largeDict.keys():
        if reviewer2 in largeDict.keys():
            sum = 0
            for k in largeDict[reviewer1].keys():
                if k in largeDict[reviewer2].keys():
                    sum = sum + (largeDict[reviewer1].get(k) - largeDict[reviewer2].get(k)) * (largeDict[reviewer1].get(k) - largeDict[reviewer2].get(k))
            return math.sqrt(sum)
        return -2
    return -1

try:
    moviefile = input("Give the name of the movie reviews file: ")
    fhandle = open(moviefile,'r')
    for line in fhandle:
        content = line.split(',')
        name = content[0]
        newDict = {}
        #print(name)
        for i in range(1,len(content),2):
            read_name = content[i]
            read_review = float(content[i+1])
            newDict[read_name] = read_review
            largeDict[name] = newDict
    
    while True:
        inputchoice = int(input("What do you want to do? Input 1 for similarity between two reviewers, or Input 2 for similarity between one reviewer and all others in the database or 3 to quit: "))

        if inputchoice == 1:
            reviewer1 = input("Provide Reviewer1 name: ")
            if not reviewer1:
                print("Please provide some input!")
            reviewer2 = input("Provide Reviewer2 name: ")
            if not reviewer2:
                print("Please provide some input!")

            score = similarityScore(reviewer1,reviewer2)
            if score == -1:
                print("Reviewer1 doesn't exist")
            elif score == -2:
                print("Reviewer2 doesn't exist")
            else:
                print("The similarity score between ",reviewer1, " and " ,reviewer2, " Chappell is: ",round(score,4))
        elif inputchoice == 2:
            review1 = input("Provide Reviewer name: ")
            if review1 in largeDict.keys():
                print("The Similarity Scores are: ")
                for review2 in largeDict.keys():
                    if review1 == review2:
                        continue
                    score = similarityScore(review1,review2)
                    print(review1,"\t",review2,"\t\t",format(round(score,2)))
            else:
                print("Reviewer doesn't exist")
        elif inputchoice == 3:
            print("Goodbye!")
            break
        else:
            print("Please enter only 1,2 or 3 as input values")

except:
    print('Please provide a valid Filename')
```
